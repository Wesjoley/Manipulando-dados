```python
#importei bibliotecas para manipulação
import numpy as np
import pandas as pd
```

```python


#colcar um arquivo dentro de um tabela 
df= pd.read_csv("agricultural_raw_material.csv" )
```

**explorando o dados**

```python
#mostra principais informacoes da tabela
df.info
```

```python
#cerificar se tem registro nulos
df.isnull().sum()
```

```python
# Trocando os dados que estao preenchidos para vazios , dados errados ou invalidos
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)
```

```python
#verifiando de novo
df.isnull().sum()
```

```python

#apagar valores nulos 

df=df.dropna()
```

```python
#verifiando de novo
df.isnull().sum()
```

```python
# Convertendo os itpo para float por conta de valores numericos e de percentuais
lst = ["Coarse wool Price", "Coarse wool price % Change", "Copra Price", "Copra price % Change", "Cotton price % Change","Fine wool Price", "Fine wool price % Change", "Hard log price % Change", "Hard sawnwood price % Change", "Hide price % change", "Plywood price % Change", "Rubber price % Change", "Softlog price % Change", "Soft sawnwood price % Change", "Wood pulp price % Change"]
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
#trazer 5 primeiras linha da df 
df.head()
```

```python
#formatando a coluna datetime e definindo-a como índice para o conjunto de dados
df.Month  = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
# Indexing month
df = df.set_index('Month')
```


### Resumo
- **`%b`**: Converte o mês abreviado (como JAN, FEB) para seu equivalente numérico (01, 02).
- **`%y`**: Converte o ano de dois dígitos (como 21, 22) para seu equivalente completo (2021, 2022).
- **`pd.to_datetime()`**: Converte as strings formatadas para objetos datetime.
- **`set_index('Month')`**: Define a coluna `Month` (agora datetime) como o índice do DataFrame, facilitando análises baseadas em da avise!

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
# %matplotlib inline: Este comando é específico para notebooks Jupyter. Ele garante que os gráficos gerados pelo Matplotlib sejam exibidos diretamente no notebook.
%matplotlib inline

# sns.set_style('darkgrid'): Este comando define o estilo dos gráficos gerados pelo Seaborn. 'darkgrid' adiciona um fundo escuro com uma grade clara.
sns.set_style('darkgrid')

# Configurações do Matplotlib:
# matplotlib.rcParams['font.size'] = 14: Define o tamanho da fonte dos textos nos gráficos para 14.
matplotlib.rcParams['font.size'] = 14

# matplotlib.rcParams['figure.figsize'] = (9, 5): Define o tamanho padrão das figuras geradas pelo Matplotlib. Neste caso, 9 unidades de largura por 5 unidades de altura.
matplotlib.rcParams['figure.figsize'] = (9, 5)

# matplotlib.rcParams['figure.facecolor'] = '#00000000': Define a cor de fundo das figuras. '#00000000' representa uma cor preta totalmente transparente.
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# Lista de materiais brutos
raw_data = ['Coarse wool Price',  'Copra Price', 'Cotton Price', 'Fine wool Price',  'Hard log Price', 'Hard sawnwood Price',
 'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']
```

```python
# Obtendo a matriz de correlação
corrmat = df[raw_data].corr()

# Definindo o tamanho do gráfico
fig = plt.figure(figsize = (12, 9))

# Mascarando a parte superior da matriz de correlação, pois a matriz é simétrica (valores repetidos)
mask = np.triu(np.ones_like(corrmat, dtype=bool))

# Criando um mapa de calor (heatmap) da matriz de correlação
sns.heatmap(corrmat, vmax = .8, mask=mask, square = True, annot = True)

# Mostra o gráfico
plt.show()
```

```python
axes=df[["Coarse wool Price", "Coarse wool price % Change"]].plot(figsize=(11, 9), subplots=True, linewidth=1)
```

```python
#df[["Wood pulp Price", "Wood pulp price % Change"]].hist(figsize=(11, 9), linewidth=1)

changelist=['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']
for i in range(len(changelist)):
    plt.figure(figsize=(12,12))
    df[changelist[i]].hist(figsize=(11, 9), linewidth=1)
    plt.xlabel('% Change')
    plt.ylabel('count')
    plt.legend(changelist[i:],loc='upper center',bbox_to_anchor=(1.2,1))
```
